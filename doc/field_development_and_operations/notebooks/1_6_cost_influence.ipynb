## Figure — Front-end loading: ability to influence cost vs cost of changes across the project lifecycle.

Produces ``1_6_cost_influence_curve.png``.


```python
import os
import matplotlib.pyplot as plt

os.makedirs("../figures", exist_ok=True)
phases = ["Exploration", "Appraisal", "Concept", "FEED", "Execution",
          "Operation"]
share = [1, 3, 5, 11, 65, 15]
colors = ["#cfe2f3", "#cfe2f3", "#fce5cd", "#fce5cd", "#d97706", "#d9ead3"]

fig, ax = plt.subplots(figsize=(8.5, 4.5))
bars = ax.bar(phases, share, color=colors, edgecolor="#333")
for b, v in zip(bars, share):
    ax.text(b.get_x() + b.get_width() / 2, b.get_height() + 1.2,
            f"{v}%", ha="center", fontsize=10)
ax.set_ylabel("Share of total project spend (%)")
ax.set_ylim(0, 80)
ax.set_title("Front-end loading: ability to influence cost vs cost of changes across the project lifecycle.", fontsize=11)
ax.grid(axis="y", alpha=0.3)
ax.spines["top"].set_visible(False); ax.spines["right"].set_visible(False)
fig.tight_layout()
fig.savefig("../figures/1_6_cost_influence_curve.png", dpi=150, bbox_inches="tight")
plt.close(fig)
print("wrote ../figures/1_6_cost_influence_curve.png")
```
